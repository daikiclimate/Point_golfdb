```python
def img2event(base_img, img):
    img3 = base_img.astype(float) - img.astype(float)
    #print(img3)
    th = 10

    index1 = img3 > th
    img3[index1] = 128

    index2 = img3 < -1 * th
    img3[index2] = 255

    img3[~(index1) & ~(index2)] = 0
    return img3.astype(np.uint8)
```

```python
def img2event(base_img, img, th = 10):
    img3 = base_img.astype(float) - img.astype(float)
    positive = np.where(img3 > th)
    positive_points = np.array([[positive[0][i], positive[1][i],
                                 positive[2][i], 1] for i in range(len(positive[0]))])
    
    negative = np.where(img3 < -1 * th)
    negative_points = np.array([[negative[0][i], negative[1][i],
                                 negative[2][i], -1] for i in range(len(negative[0]))])

    return np.concatenate([positive_points, negative_points], axis = 0)
img2event(img1, img2)
```

```python
dim = 160
path ="/home/ubuntu/datasets/golf/videos_160/578.mp4"

cap = cv2.VideoCapture(path)
fourcc = cv2.VideoWriter_fourcc(*"mp4v")
out = cv2.VideoWriter("sample.mp4",             fourcc, cap.get(cv2.CAP_PROP_FPS), (dim, dim))

count = 0
success, image = cap.read()
#image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
base_image = image[:]
while success:
    count += 1
    b_img = img2event(base_image.copy(), image.copy())

    out.write(b_img)
    base_image = image[:]

    success, image = cap.read()
    #image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    
cap.release()
cv2.destroyAllWindows()
```
